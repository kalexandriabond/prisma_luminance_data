```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

from jupyterthemes import jtplot
```

```python
jtplot.style(theme='grade3', context='poster', spines=False)
```

```python
home = os.path.expanduser('~')
```

```python
lum_data_path = home + '/Dropbox/loki_0.5/measuring_luminance/'
lum_100p_df = pd.read_csv(lum_data_path + '100p_lum_data.csv')
lum_50p_df = pd.read_csv(lum_data_path + '50p_lum_data.csv')
lum_25p_df = pd.read_csv(lum_data_path + '25p_lum_data.csv')
```

```python
lum_100p_df.head(), lum_50p_df.head(), lum_25p_df.head()
```

```python
lum_levels = np.arange(20)
```

```python
bw_lighting_dict = dict({'100': lum_100p_df.bw, '50': lum_50p_df.bw, 
                         '25': lum_25p_df.bw, 'lum_level': lum_levels})
bw_lighting_df = pd.melt(pd.DataFrame(bw_lighting_dict), id_vars='lum_level',var_name='light', value_name='lum_obs')
```

```python
red_lighting_dict = dict({'100': lum_100p_df.red, '50': lum_50p_df.red,
                          '25': lum_25p_df.red,'lum_level': lum_levels})
red_lighting_df = pd.melt(pd.DataFrame(red_lighting_dict), id_vars='lum_level',var_name='light', value_name='lum_obs')
```

```python
green_lighting_dict = dict({'100': lum_100p_df.green, '50': lum_50p_df.green, 
                            '25': lum_25p_df.green,'lum_level': lum_levels})
green_lighting_df = pd.melt(pd.DataFrame(green_lighting_dict),  id_vars='lum_level', var_name='light', value_name='lum_obs')
```

```python
blue_lighting_dict = dict({'100': lum_100p_df.blue, '50': lum_50p_df.blue,
                           '25': lum_25p_df.blue, 'lum_level': lum_levels})
blue_lighting_df = pd.melt(pd.DataFrame(blue_lighting_dict), id_vars='lum_level', var_name='light', value_name='lum_obs')
```

```python
sns.pointplot(x='lum_level', y='lum_obs', hue='light', data=red_lighting_df, 
              color='red', hue_order=['25','50','100'])
plt.title('Red')
plt.ylabel(r'cd/$m^2$')
plt.xlabel('luminance in psychopy units')
```

```python
sns.pointplot(x='lum_level', y='lum_obs', hue='light', data=green_lighting_df, 
              color='green',hue_order=['25','50','100'])
plt.title('Green')
plt.ylabel(r'cd/$m^2$')
plt.xlabel('luminance in psychopy units')
```

```python
sns.pointplot(x='lum_level', y='lum_obs', hue='light', 
              data=blue_lighting_df, color='blue', hue_order=['25','50','100'])
plt.title('Blue')
plt.ylabel(r'cd/$m^2$')
plt.xlabel('luminance in psychopy units')
```

```python
sns.pointplot(x='lum_level', y='lum_obs', hue='light', data=bw_lighting_df, 
              color='gray', hue_order=['25','50','100'])
plt.title('BW')
plt.ylabel(r'cd/$m^2$')
plt.xlabel('luminance in psychopy units')
```
